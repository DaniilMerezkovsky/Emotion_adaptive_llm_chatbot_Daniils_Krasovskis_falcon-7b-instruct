```python
from transformers import pipeline

# Loading LLM 4 min
llm = pipeline(
    "text-generation",
    model="tiiuae/falcon-7b-instruct",
    device_map="auto",
    max_new_tokens=150,
    temperature=0.7,
)

```

```python
# Generate function
def chat(prompt):
    response = llm(prompt)[0]["generated_text"]
    print("User:", prompt)
    print("Model:", response)
    print("-" * 80)


# basic emotions
prompts = [
    "I am so angry right now.", # anger
    "I feel deep sadness today.", # sadness
    "I am terrified about tomorrow.", # fear
    "That rotten smell made me feel disgusted.", # disgust
    "I just got accepted into university and I'm so happy!", # joy

    # complex emotions
    "My best friend achieved something I wanted for myself, and now I feel jealous and ashamed of it.", #jealousy
    "I lied to someone I care about, and now I feel guilty even though I thought I was helping.", # guilt
    "I found an old photo from childhood, and it makes me feel warm and sad at the same time.", # nostalgia
    "I promised to keep a secret, but now telling the truth might save someone from getting hurt.", # moral conflict
    "Sometimes I think about how small I am compared to the universe, and it fills me with both fear and fascination." # existential awe
]

for p in prompts:
    chat(p)
```
